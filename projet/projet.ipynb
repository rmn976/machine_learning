```python
import warnings
import pandas as pd 
import numpy as np 
import seaborn as sns
import sklearn as sklearn
import matplotlib.pyplot as plt
%matplotlib inline
warnings.simplefilter('ignore')
```

```python
# Affichage du dataset
df = pd.read_csv('train.csv', sep=';'); df.head()
```

```python
# Taille du dataset
df.shape
```

```python
# Remplacement des donn√©es binaire par 0 et 1
df['y'].replace(('yes', 'no'), (1, 0), inplace=True)
df['default'].replace(('yes', 'no'), (1, 0), inplace=True)
df['housing'].replace(('yes', 'no'), (1, 0), inplace=True)
df['loan'].replace(('yes', 'no'), (1, 0), inplace=True)
```

```python
# Encodage des donn√©es cat√©gorielles avec sklearn encoder
from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
df['job'] = lb.fit_transform(df['job'])
df['marital'] = lb.fit_transform(df['marital'])
df['education'] = lb.fit_transform(df['education'])
df['contact'] = lb.fit_transform(df['contact'])
df['month'] = lb.fit_transform(df['month'])
df['poutcome'] = lb.fit_transform(df['poutcome'])
df.head()
```

```python
# Statistiques descriptives du dataset
df.describe()
```

```python
# Visualisation en bo√Æte √† moustache de la feature 'poutcome'
plt.figure(figsize=(16, 5))
sns.boxplot(x="poutcome", data=df)
```

```python
# Exemple de donn√©es retir√©es du dataset
over70 = df['age'] > 70
df_over70 = df[over70]
df_over70.shape
```

```python
# Suppression des valeurs ab√©rantes repr√©sentant moins de 5% du dataset
filter1 = df['age'] < 70
df = df[filter1]
filter2 = df['balance'] < 3000
df = df[filter2]
filter3 = df['balance'] > -1500
df = df[filter3]
filter4 = df['duration'] < 800
df = df[filter4]
filter5 = df['campaign'] < 8
df = df[filter5]
print(df.shape)
```

```python
# Heatmap des corr√©lations
f, ax = plt.subplots(figsize=(20,7))
sns.heatmap(df.iloc[:,-15:].corr(), annot=True, ax =ax)
```

```python
# On d√©finit X avec les features en entr√©e
X = df[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome']]
X.head()
```

```python
# On d√©finit y avec la classe servant de r√©sultat
y = df[['y']]
y.shape
```

```python
# On split nos datas pour avoir des set d'entrainement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Algo 1

```python
# On cr√©e notre classifier Random Forest avec 2 param√®tres
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(n_estimators= 100, max_depth= 50)

# On entraine notre model
clf1.fit(X_train, y_train)
y_pred = clf1.predict(X_test)
```

```python
# Accuracy de notre classifier
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import metrics
from pandas import DataFrame
import matplotlib.pyplot as plt 
prob_score = clf1.predict_proba(X_test)
prob_score = DataFrame(prob_score).iloc[:,0]
fpr,tpr,thresholds = metrics.roc_curve(y_test,1-prob_score)
```

```python
test_fpr, test_tpr, _ = metrics.roc_curve(y_test.values, y_pred)
test_auc = np.round(metrics.auc(test_fpr, test_tpr), 3)
```

```python
plt.plot(test_fpr,test_tpr, color = 'darkorange')
plt.xlim([-.05, 1.05])
plt.ylim([-.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.show()
```

```python
test_fpr = np.insert(test_fpr, 0, 0)
test_tpr = np.insert(test_tpr, 0, 0)

fig, ax = plt.subplots(figsize=(8, 6))
#ax.plot(train_fpr, train_tpr, label='Train AUC: ' + str(train_auc))
ax.plot(test_fpr, test_tpr, label='Test AUC: ' + str(test_auc))
#ax.plot(test_fpr, test_tpr, 'k--', label='Chance Curve')
ax.set_xlabel('False Positive Rate', fontsize=12)
ax.set_ylabel('True Positive Rate', fontsize=12)
ax.grid(True)
ax.legend(fontsize=12)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = clf1.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(cm).plot()
```

```python
importances = clf1.feature_importances_
importances
```

```python
feature_imp = pd.Series(importances, index=X.columns.tolist()).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 9))
sns.barplot(x=feature_imp, y=feature_imp.index)
```
